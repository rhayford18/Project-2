```python
class ElectronicDevice:
    '''This is the superclass'''
    def __init__(self,brand):
        self.brand = brand 
        self._on = False 

    def power_on(self):
        print(f"Turning on the {self.brand} device")
        self._on = True
        print("Device is now on")

    
class Camera(ElectronicDevice):
    '''Camera subclass'''
    def take_photo(self):
        print(f"Trying to take a photo on the {self.brand} device")
        if self._on:
            print(("> Click! Photo saved to gallery."))
        else:
            print("> Device is not on.")

class Phone(ElectronicDevice):
    '''Phone subclass'''
    def power_on(self):
        super().power_on()
        print("connected to a celluar network")

    def call(self, phone_number):
        print(f"Trying to make a call on the {self.brand} device")
        if self._on:
            print(f"> Dialed{phone_number}.")
        else:
            print("> Device is not on")


'''Printing classes'''
camera = Camera("Sony")
camera.take_photo()
camera.power_on()
camera.take_photo()
print()

phone = Phone("Nokia")
phone.call("555-555-5555")
phone.power_on()
phone.call(" 555-555-5555")
```

```python
class Smartphone(Phone, Camera):
    def __init__(self, brand, os):
        super().__init__(brand)
        self.os = os
    
    def power_on(self):
        super().power_on()
        print(f" loaded {self.os} operating system")

'''Printing Smartphone class'''
smartphone = Smartphone("Apple", "iOS")
smartphone.take_photo()
smartphone.power_on()
smartphone.take_photo()
smartphone.call(" 573-333-1234")
```
