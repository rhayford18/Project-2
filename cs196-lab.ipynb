```python
class Point:
    '''The Point class '''
    def __init__(self, x, y):
       self.x = x
       self.y = y 
    def __repr__(self):
        return f"Point({self.x}, {self.y})"
    def __add__(self,other):
        if isinstance(other,Point):
            return Point(self.x + other.x, self.y +other.y)
    def __eq__(self, other):
        return self.x == other.x and self.y == other.y
    def distance_to_origin(self, other):
            return (self.x**2 + self.y**2)**0.5
```

```python
class Rect:
    def __init__(self, x, y, width, height):
        '''This is the initation of the class '''
        self.x = x
        self.y = y 
        self.width = width 
        self.height = height 
    def __repr__(self):
        '''This shows how the text will come when printed'''
        return f"Retangle(x = {self.x}, y = {self.y}, width = {self.width}, height = {self.height})"
    def move(self, new_x, new_y):
        '''moves the x and y value to whatever the user puts later on'''
        self.x = new_x
        self.y = new_y 

    def flip(self):
        #This will make it so that the height and width can be flipped
        self.width, self.height = self.height, self.width
    def __contains__(self, other):
        assert isinstance(other, (Point, Rect)), "Input must be a point or Rect"

        if isinstance(other, Point):
            return (self.x <= other.x <= self.x + self.width and self.y <= other.y <= self.y +self.height)
        
        if isinstance(other, Point):
            if self.x + self.width < other: return False
            if self.x > other.x + other.width: return False
            if self.y + self.height < other: return False 
            if self.y > other.y + other.height: return False
            return True
        
    def area(self):
        return self.width * self.height
```
