```python
from collections import namedtuple

Point = namedtuple('point',('x','y'))
p = Point(2, 3)


```

```python
from dataclasses import dataclass, field
import random 

def createStudentId() -> int:
    return random.randint(1, 1000)

@dataclass(order=True)
class Rect:
    id: int = field(default_factory= createStudentId, init=False)
    x: int 
    y: int 
    width: int
    height: int|None = None 
    def post__init__ (self):
        if self.height is None:
            self.height = self.width




```

```python
p1 = Point(5, 2)
p2 = Point(6, 7)
p3 = p1+p2

print(p1)
print(p2)
print(p3)
```

```python
r1 = Rect(0,0,10,20)
r2 = Rect(5,5,15,25)
r3 = Rect(10,10,12)
r4 = Rect(20,20,8)

rectangles = [r1, r2, r3, r4]

print(sorted(rectangles))
```
